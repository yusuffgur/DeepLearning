**Plot the function graph in a new notebook. Use PyTorch for activation functions.**

```python
%matplotlib inline
import torch
import numpy as np
import matplotlib.pylab as plt
from matplotlib import pyplot as plt
from matplotlib import style
from IPython import display
import os
```

```python
checkFolder=os.path.exists('activation')
if not (checkFolder):
  os.makedirs('activation')
```

```python
def activationFunctionPlot(x, y, name):
    display.set_matplotlib_formats('svg')
    fig = plt.figure() #Here is your err
    plt.rcParams['figure.figsize'] = (8, 5)
    plt.plot(x.detach().numpy(), y.detach().numpy(), label = name, linewidth=1.5, color='#0000FF')
    plt.grid(True,linestyle=':')   
    plt.legend(loc='upper left') 
    plt.style.use("seaborn")

    ax = plt.gca()
    ax.spines['right'].set_color("none")
    ax.spines['top'].set_color("none")
    ax.spines['bottom'].set_position(("data",0))  
    ax.spines['left'].set_position(("data",0)) 
    ax.spines['bottom'].set_linewidth(0.5)
    ax.spines['left'].set_linewidth(0.5)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    plt.savefig('./activation/'+name+'_activationFunction.png')
```

```python
#sigmoidActivation function
x = torch.arange(-6.0, 6.0, 0.01, requires_grad=True)
y = x.sigmoid()
activationFunctionPlot(x, y, 'sigmoid')
```

```python
#tanhActivation function
x = torch.arange(-6.0, 6.0, 0.01, requires_grad=True)
y = x.tanh()
activationFunctionPlot(x, y, 'tanh')
```

```python
#reluActivation function
x = torch.arange(-6.0, 6.0, 0.01, requires_grad=True)
y = x.relu()
activationFunctionPlot(x, y, 'relu')
```

```python
#softmaxActivation function

x = torch.arange(-6.0, 6.0, 0.01, requires_grad=True)
y = x.softmax(dim=0)
activationFunctionPlot(x, y, 'softmax')
```
